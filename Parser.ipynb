***TAGGING VIA PYCORENLP WRAPPER LIBRARY***

```python
from pycorenlp import StanfordCoreNLP
nlp = StanfordCoreNLP('http://localhost:9000')
doc = "The King of France is bald."
annot_doc = nlp.annotate(doc, properties={'annotators':'pos', 'outputFormat':'json'})
annot_doc
```

```python
for sent in annot_doc["sentences"]:
    for word in sent["tokens"]:
        print(f"{word['word']} -> {word['pos']}")
```

```python
def tagCorpus(sheet):
    tags = []
    for sent in sheet["Item"]:
        tags.append(tagSents(sent))
    return tags
def tagSents(item):
    doc = nlp.annotate(item, properties={'annotators':'pos', 'outputFormat':'json'})
    toks = []
    for sent in doc["sentences"]:
        for word in sent["tokens"]:
            toks.append({word['word']: word['pos']})
    return toks
tags = tagCorpus(stimuli_sheet)
```
